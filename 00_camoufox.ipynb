```python
# Cell 1: 启动浏览器
from camoufox.async_api import AsyncCamoufox

# 创建浏览器实例（全局变量，供后续 cell 使用）
camoufox = AsyncCamoufox(headless=False)  # headless=False 显示浏览器窗口
browser = await camoufox.__aenter__()
page = await browser.new_page()

print("✅ 浏览器已启动")
```

```python
# Cell 2: 打开网页
await page.goto("https://example.com")
print(f"📄 页面标题: {await page.title()}")
print(f"🔗 当前URL: {page.url}")
```

```python
# Cell 3: 截图保存
await page.screenshot(path="screenshot.png", full_page=True)
print("📸 截图已保存到 screenshot.png")
```

```python
# Cell 4: 获取页面内容
content = await page.content()
print(f"📝 页面HTML长度: {len(content)} 字符")

# 获取页面中的所有链接
links = await page.locator("a").all()
print(f"🔗 找到 {len(links)} 个链接")

for i, link in enumerate(links[:5]):  # 只显示前5个
    text = await link.text_content()
    href = await link.get_attribute("href")
    print(f"   {i+1}. {text} -> {href}")
```

```python
# Cell 5: 点击链接
# 点击第一个链接
if links:
    await links[0].click()
    await page.wait_for_load_state("networkidle")
    print(f"🖱️ 点击链接后，新页面: {await page.title()}")
    print(f"🔗 新URL: {page.url}")
```

```python
# Cell 6: 返回上一页
await page.go_back()
await page.wait_for_load_state("networkidle")
print(f"⬅️ 返回后页面: {await page.title()}")
```

```python
# Cell 7: 关闭浏览器
await camoufox.__aexit__(None, None, None)
print("🔒 浏览器已关闭")
```
